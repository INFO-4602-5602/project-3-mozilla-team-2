```python
import pandas as pd
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
import math
import operator
```

```python
df = pd.read_csv("rawData.csv", encoding='cp1252', low_memory=False)
df.head()
```

```python
list(df.columns.values)
```

```python
df = df.fillna(0)
```

```python
for i in range(len(df.columns)):
    print(df.columns[i] + " (" + str(i) + ")")
    print(df[df.columns[i]])
    print("\n")
```

```python
entries = len(df)
iot = range(8,19)
defns = range(34, 44)
for col in iot:
    for row in range(entries):
        if(df.iat[row, col] != 0):
            #df.iloc[row][column] = 1
            df.iat[row, col] = 1
for col in defns:
    for row in range(entries):
        if(df.iat[row, col] != 0):
            df.iat[row, col] = 1
```
